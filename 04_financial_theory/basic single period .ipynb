# 0. Concepts and Definitions

#### Variables in one-period discrete time

There are $N+1$ securities, with prices $S^{(i)}_t(\omega_j)$

- $i \in \{0,\ldots,N\}$  
- $t \in \{0,1\}$  
- $\omega_j$ is the state of the world, $j \in \{1,\ldots,m\}$  
- $p_j$ is the true probability of state $\omega_j$

In the one-period setting, $S^{(i)}_1(\omega_j)$ is often called the payoff.  
Technically, this represents the liquidation value of the asset at $t=1$, which includes the market price plus any dividends paid during the period.

In multiperiod models, $S^{(i)}_t(\omega)$ is called the **price process** (more precisely, the ex-dividend price process) of asset $i$.

The payoff terminology is no longer used for $S_t$. Instead we distinguish:

- **Price process:** $S^{(i)}_t$  
- **Dividend process:** $D^{(i)}_t$  
- **Gain (value) process:**  
  $$
  G^{(i)}_t = S^{(i)}_t + \sum_{u=1}^{t} D^{(i)}_u
  $$

----

#### **LOOP** 
Law of One Price is the most basic statement: identical payoffs must have identical prices. 

#### **No-arbitrage** 
The economic/financial principle that enforces the law of one price. If two identical payoffs had different prices, you could exploit the gap for a free profit, which is the arbitrage. The LOOP follows from no-arbitrage.

#### **Linear pricing** 
A consequence of no-arbitrage. Once you accept no-arbitrage, you get the law of one price, and then linear pricing follows as <span style="color:red">its natural extension to portfolios</span>: not just "same payoff, same price" but "scaled and combined payoffs scale and combine in prices too."

#### **FTAP** (Fundamental Theorem of Asset Pricing)
No-arbitrage is equivalent to the <span style="color:red">existence of a EMM / risk-neutral measure (or a state price vector / stochastic discount factor, depending on the setting)</span>. Gives you the full mathematical structure that no-arbitrage implies, and linear pricing drops out immediately from that structure because expectation operators are linear.


### 01. Types of arbitrage
* Type A: Immediate profit without any futre cost. Ex. someone gives you money. 
* Type B: non-positve cost, but with positive probability of producing a positive payoff in future and zero prob o producing a negative payoff.

In one period dicrete setting, $\theta$ is a trading strategy:
* Type A arbitrage: 
$$
S_0^\top \theta < 0 \quad \text{and} \quad S_1 \theta = 0.
$$

* Type B arbitrage:
$$
S_0^\top \theta \le 0, \quad S_1 (\omega_j) \theta \ge 0, \quad \text{and} \quad S_1 \theta \ne 0.
$$

Note here that:
1. $\quad S_1 (\omega_j) \theta \ge 0, \quad$: the portfolio payoff is never negative in any state
2. $\quad S_1 \theta \ne 0$: there is at least one state where the payoff is strictly positive.


### 02. Elementary security

A security that pays 1 in state of the world and zero on all other states, with the payoff of the form

$e_j = (0, 0, 0, 1, 0...0)$ when state is $\omega_j$.

Other names (aka):
* Arrow-Debreu security (named after Kenneth Arrow and Gerard Debreu from their 1950s general equilibrium work)
* pure security
* state-contingent claim

Note that for $m$ states there are possibly $m$ elementary securities. We will see later that when there are all elementary securities this is a complete market.

### 03. Attainable Security

In discrete time, single period, a security or contingent claim $X$ is **attainable** 
if there exists a trading strategy $\theta$ that is a replicating portfolio.

$$
\begin{pmatrix}
X(\omega_1) \\
\vdots \\
X(\omega_m)
\end{pmatrix}
=
\begin{pmatrix}
S^{(0)}_1(\omega_1) & \cdots & S^{(N)}_1(\omega_1) \\
\vdots & \ddots & \vdots \\
S^{(0)}_1(\omega_m) & \cdots & S^{(N)}_1(\omega_m)
\end{pmatrix}
\begin{pmatrix}
\theta_0 \\
\vdots \\
\theta_N
\end{pmatrix}
$$

In what comes we will consider:
* $m$ possible states of the world $\omega_k$, each with probability $p_k$
* $N+1$ securities, $S^{(i)}_t$, where $i$ indexes the security and $t \in \{0,1\}$

In short: $X = S_1 \theta$


Note: the term SECURITY is used as synonim of CONTINGENT CLAIM. Not only calls/puts, etc... bcs securities ARE contingent CLAIMS!!! Their values depends (are contingent) on the states of the world.

#### 04. The State Price Vector $\pi$

The state price vector $\pi \in \mathbb{R}^m$ it the one that can be used to recover the price $P$ of any attainable security $X$ at time $t=0$:

$$
P = \sum_{k=1}^{K} \pi_k X(\omega_k).
$$

Notice the state-price vector has one entry per state:

$$\pi = \begin{pmatrix} \pi_1 \\ \vdots \\ \pi_m \end{pmatrix}$$

and that $\pi_k$ is the price today of an Arrow-Debreu security that pays $1$ in state 
$\omega_k$ and $0$ in all other states.

or in matrix form:

$$S_0 = X^\top \pi$$


### 04. Cash Account

A security that earns the risk-free interest rate prevailing at the time of deposit, 
in every state of the world. When available, it is denoted $S^{(0)}_t$ and is defined as:

$$S^{(0)}_0 = 1, \qquad S^{(0)}_1 = 1 + r$$

### 05. Numeraire and Deflated Prices
Any security with strictly postive price processes can serve as a numeraire security. What is means to be a numeraire security? Means writing the other price processes in another unit of account (number of numeraire securities).


Let $S^{(n)}$ be the numeraire security. We redefine all prices as:

$$\tilde{S}^{(i)}_t = \frac{S^{(i)}_t}{S^{(n)}_t}$$

where $\tilde{S}^{(i)}_t$ is the **deflated price** of security $i$ at time $t$.

By construction, the numeraire always has deflated price equal to 1:

$$\tilde{S}^{(n)}_t = \frac{S^{(n)}_t}{S^{(n)}_t} = 1$$


---

### Risk-neutral Probability Measure or Equivalent Martingale Measures (EMMs)

We assume a chosen numeraire security with price process $S^{(n)}_t$.

---

### Def 1: Deflated Prices

For any security $i$, define its **deflated price** at time $t$ as:

$$\tilde{S}^{(i)}_t := \frac{S^{(i)}_t}{S^{(n)}_t}$$

This is just the price of security $i$ measured in units of the numeraire, 
rather than in dollars.

---

### Def 2: Martingales

A process $Y_t$ is a **martingale** under a probability measure $\mathbb{Q}$ if 
your best forecast today of its future value is simply its current value:

$$\mathbb{E}^{\mathbb{Q}}_0[Y_1] = Y_0$$

There is no expected drift up or down. The process is, on average, flat.

---

### Expectation Under $\mathbb{Q}$

Given probabilities $\mathbb{Q} = (q_1, \ldots, q_m)$ over states 
$\omega_1, \ldots, \omega_m$, the expectation of any payoff $X$ is:

$$\mathbb{E}^{\mathbb{Q}}_0[X] = \sum_{k=1}^{m} q_k \, X(\omega_k)$$

---

### The EMM Definition (using 1, 2, 3)

An **equivalent martingale measure (EMM)**, or 
**risk-neutral probability measure**, is a set of probabilities 
$\mathbb{Q} = (q_1, \ldots, q_m)$ such that:

1. $q_k > 0$ for all $k$

2. Every deflated price is a martingale under $\mathbb{Q}$:

$$\mathbb{E}^{\mathbb{Q}}_0\left[\tilde{S}^{(i)}_1\right] = \tilde{S}^{(i)}_0 \quad \text{for all } i$$

Expanding using the definition of deflated prices from Step 1:

$$\mathbb{E}^{\mathbb{Q}}_0\left[\frac{S^{(i)}_1}{S^{(n)}_1}\right] = \frac{S^{(i)}_0}{S^{(n)}_0}$$

These $q_k$ are not necessarily the true real-world probabilities. They are 
a mathematical construction that will make deflated prices behave like martingales.
The EMM $\mathbb{Q}$ is tied to the chosen numeraire $S^{(n)}_t$. A different 
numeraire gives a different EMM. It is therefore more precise to speak of an 
**EMM-numeraire pair** rather than the EMM alone.

# 1. Main results: martingale pricing theory in single period models

1. If EMM $\mathbb{Q}$ exists there can be NO arbitrage opportunities




2. If there is a security $S_0^{(n)}$ with strictly poistive price process, and $\pi_k$ state price vectors, then there is a EMM (aka risk newutral measure).

Proof: Starting from state-price representation:

$$
S_0^{(j)} = \sum_{k=1}^m \pi_k S_1^{(j)}(\omega_k)
$$

Multiply and divide by the numeraire and its normalization in one step:

$$
S_0^{(j)}
=
\sum_{k=1}^m
\frac{\pi_k S_1^{(n)}(\omega_k)}{\sum_{l=1}^m \pi_l S_1^{(n)}(\omega_l)}
\cdot
\frac{S_1^{(j)}(\omega_k)}{S_1^{(n)}(\omega_k)}
\cdot
\sum_{l=1}^m \pi_l S_1^{(n)}(\omega_l)
$$

Recognize:
$$
\quad
q_k = \frac{\pi_k S_1^{(n)}(\omega_k)}{S_0^{(n)}}, \\ \\ \\
S_0^{(n)} = \sum_{l=1}^m \pi_l S_1^{(n)}(\omega_l)

$$

Hence:
$$
\frac{S_0^{(j)}}{S_0^{(n)}} =
\sum_{k=1}^m q_k \frac{S_1^{(j)}(\omega_k)}{S_1^{(n)}(\omega_k)}
= \mathbb{E}^\mathbb{Q}\!\left[\frac{S_1^{(j)}}{S_1^{(n)}}\right]
$$

3. Existence of EMM = Existence of positive state prices = NO arbitrage

4. Unique EMM = complete markets. For one period, no arbitrage means $\text{rank} (S_1)=m$.
