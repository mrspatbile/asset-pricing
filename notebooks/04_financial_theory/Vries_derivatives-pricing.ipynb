# Random walk

A discrete time, discrete space stochastic process $\{S_t\}$ defined as:

$$S_t = S_0 + \sum_{i=1}^{t} \epsilon_i$$

where $\epsilon_i \overset{iid}{\sim} (0, \sigma^2)$ are independent, identically distributed increments.

#### Properties

- **Mean**: $\mathbb{E}[S_t] = S_0$ (martingale)
- **Variance prop to time**: $\text{Var}(S_t) = \sigma^2 t$
- **Covariance**: $\text{Cov}(S_t, S_s) = \sigma^2 \min(t, s)$
- **Independent Increments**: $S_t - S_s \perp S_s - S_0$ for $t > s > 0$

#### Simple Symmetric Random Walk

$$\epsilon_i = \begin{cases} +1 & \text{with probability } \frac{1}{2} \\ -1 & \text{with probability } \frac{1}{2} \end{cases}$$

So $\mathbb{E}[\epsilon_i] = 0$ and $\text{Var}(\epsilon_i) = 1$.

#### Scaling Limit (Donsker's Theorem)

As $n \to \infty$:

$$\frac{S_{\lfloor nt \rfloor}}{\sqrt{n}} \xrightarrow{d} W_t$$

where $W_t$ is a standard Brownian motion.

#### Random Walk with Drift

$$S_t = S_0 + \mu t + \sigma \sum_{i=1}^{t} \epsilon_i$$

- **Mean**: $\mathbb{E}[S_t] = S_0 + \mu t$
- **Variance**: $\text{Var}(S_t) = \sigma^2 t$



#### Quadratic Variation

$$[S]_t = \sum_{i=1}^{t} (\Delta S_i)^2 = \sigma^2 t$$

This is deterministic, a key distinction from diffusion processes.


````python
# Parameters
n_steps = 10000
n_paths = 1000
p = 0.5
seed = 123

rng = np.random.default_rng(seed)

# A matrix of shape (n_paths, n_steps) with each random outcome being either +1 or -1
Z = rng.choice([1, -1], size=(n_paths, n_steps), p=[p, 1-p])

# matrix for the paths
M = np.zeros((n_paths, n_steps + 1))

# first column is zero, then cumulative sum of Z along the rows
M[:, 1:] = np.cumsum(Z, axis=1)

fig, ax = plt.subplots(figsize=(12, 6))

for i in range(n_paths):
    ax.plot(M[i], color="blue", alpha=0.2, linewidth=1.2, label=f"Path {i+1}")

ax.set_title(f"Standard Random Walk -- {n_paths} paths, {n_steps} steps, p={p}", fontsize=18)
ax.set_xlabel("Time step $i$", fontsize=12)
ax.set_ylabel("$X_i$", fontsize=12)

plt.tight_layout()
````

## Test of some ppts of random walks


#### 1. E[Mi] = 0

````python
steps = np.arange(n_steps + 1) 
std = np.sqrt(steps) / np.sqrt(n_paths)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(np.mean(M, axis=0), color="blue", alpha=0.7, linewidth=1.2)
ax.axhline(0, color="red", linewidth=0.8, linestyle="--", alpha=0.5)
for k, alpha in zip([1, 2, 3], [0.4, 0.3, 0.2]):
    ax.fill_between(steps, -k*std, k*std, alpha=alpha, color="red", label=f"{k}σ band")

ax.set_title(f"Average Mi values -- {n_paths} paths, {n_steps} steps, p={p}", fontsize=18)
ax.set_xlabel("Time step $i$", fontsize=12)
ax.set_ylabel("$Average M_i$", fontsize=12)
ax.grid(True, alpha=0.5)
ax.xaxis.grid(False)
ax.legend(fontsize=10)
for spine in [s for s in ax.spines.values() if s != ax.spines['bottom']]:
    spine.set_visible(False)
````

### 2. Martingale property



````python
i = 2000
j_range = np.arange(i, n_steps + 1)
values_at_i = M[:, i]

n_bins = 6
percentiles = np.linspace(0, 100, n_bins + 1)
finite_edges = np.percentile(values_at_i, percentiles)
bin_edges = finite_edges.copy()
bin_edges[0] = -np.inf
bin_edges[-1] = np.inf

labels = [f"{finite_edges[k]:.1f} to {finite_edges[k+1]:.1f}" for k in range(n_bins)]
labels[0] = f"< {finite_edges[1]:.1f}"
labels[-1] = f"> {finite_edges[-2]:.1f}"

colors = plt.cm.Blues(np.linspace(0.3, 0.9, n_bins))

fig, ax = plt.subplots(figsize=(12, 6))

for k in range(n_bins):
    mask = (values_at_i >= bin_edges[k]) & (values_at_i < bin_edges[k+1])
    if mask.sum() > 10:
        group_mean = np.mean(M[mask, i:], axis=0)
        ax.plot(j_range, group_mean, color=colors[k], linewidth=2, label=labels[k])

ax.axvline(i, color="black", linewidth=1, linestyle="--", alpha=0.5, label=f"Conditioning time i={i}")
ax.set_title("Martingale property: $\\mathbb{E}[M_j | M_i] = M_i$", fontsize=16)
ax.set_xlabel("Time step $j$", fontsize=12)
ax.set_ylabel("Conditional mean $M_j$", fontsize=12)
ax.grid(True, alpha=0.3)
ax.xaxis.grid(False)
for spine in [s for s in ax.spines.values() if s != ax.spines['bottom']]:
    spine.set_visible(False)
ax.legend(fontsize=10, bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
````

````python
def random_walk_2d(n_steps):
    angles = np.random.uniform(0, 2 * np.pi, size=n_steps)
    x = np.cumsum(np.cos(angles))
    y = np.cumsum(np.sin(angles))
    x = np.insert(x, 0, 0)
    y = np.insert(y, 0, 0)
    return x, y

n=1000
x, y = random_walk_2d(n)

plt.figure(figsize=(6, 6))
plt.plot(x, y, color='steelblue', linewidth=0.5, alpha=0.7)
plt.scatter(x[0], y[0], color='green', zorder=5, label='Start')
plt.scatter(x[-1], y[-1], color='red', zorder=5, label='End')
plt.title('2D Random Walk')
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.show()
````

## Geometric random walk 
For a random walk to represent reasonably stock prices it need 2 additional features:
* Add drift (bias) and volatility (variable response to a shock)
* Do not take negative numbers

GRW : 
* model log price as a linear transform of a random walk (deterministic shifting and scaling)
* model price as exponential of a linear transform of a random walk

$$S_i = S_0 e^{X_{i-1} + \mu + \sigma Z_i} = S_{i-1} e^{\mu + \sigma Z_i}$$

````python
# Parameters
n_steps = 1200
n_paths = 10
n_per_year = 100
n_years = n_steps // n_per_year  # 12
months_per_year = 12
steps_per_month = n_per_year // months_per_year  # ~8

mu_annual = 0.05
sigma_annual = 0.20
delta = 1 / n_per_year
mu = mu_annual * delta
sigma = sigma_annual * np.sqrt(delta)

p = 0.5
S0 = 100
seed = 42

rng = np.random.default_rng(seed)
Z = rng.choice([1, -1], size=(n_paths, n_steps), p=[p, 1-p])

S = np.zeros((n_paths, n_steps + 1))
S[:, 0] = S0
for i in range(1, n_steps + 1):
    S[:, i] = S[:, i-1] * np.exp(mu + sigma * Z[:, i-1])

colors = plt.cm.Blues(np.linspace(0.3, 0.9, n_paths))

fig, ax = plt.subplots(figsize=(14, 6))

# monthly gridlines -- light gray
for m in range(1, n_years * months_per_year):
    ax.axvline(m * steps_per_month, color="gray", linewidth=0.4, alpha=0.3, zorder=0)

# yearly gridlines -- darker gray
for y in range(1, n_years + 1):
    ax.axvline(y * n_per_year, color="gray", linewidth=1.0, alpha=0.7, zorder=0)

# paths
for i in range(n_paths):
    ax.plot(S[i], color=colors[i], linewidth=1.2, zorder=1)

ax.axhline(S0, color="red", linewidth=0.8, linestyle="--", alpha=0.5)

# x axis ticks at years only
ax.set_xticks([y * n_per_year for y in range(n_years + 1)])
ax.set_xticklabels([str(y) for y in range(n_years + 1)])

ax.set_title(f"Geometric random walk -- {n_paths} paths, {n_years} years", fontsize=16)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("$S_i$", fontsize=12)
ax.xaxis.grid(False)
ax.yaxis.grid(False)
plt.tight_layout()
````

## Scaling μ and σ from annual to per-step

### Setup
- Annual drift: $\mu_{annual} = 0.05$ (5% per year)
- Annual volatility: $\sigma_{annual} = 0.20$ (20% per year)
- Steps per year: $n = 100$
- Time step: $\delta = \frac{1}{n} = \frac{1}{100} = 0.01$ years

### Scaling rules

$$\mu_{step} = \mu_{annual} \cdot \delta = \mu_{annual} \cdot \frac{1}{n}$$

$$\sigma_{step} = \sigma_{annual} \cdot \sqrt{\delta} = \sigma_{annual} \cdot \frac{1}{\sqrt{n}}$$

Note: $\mu$ scales linearly with $\delta$, $\sigma$ scales with $\sqrt{\delta}$ -- because variance is additive, not standard deviation.

### Substitution

$$\mu_{step} = 0.05 \times \frac{1}{100} = 0.0005$$

$$\sigma_{step} = 0.20 \times \frac{1}{\sqrt{100}} = \frac{0.20}{10} = 0.02$$

### Verification: recover annual parameters

After $n = 100$ steps of length $\delta$:

$$\mu_{annual} = \mu_{step} \times n = 0.0005 \times 100 = 0.05 \checkmark$$

$$\sigma_{annual} = \sigma_{step} \times \sqrt{n} = 0.02 \times \sqrt{100} = 0.02 \times 10 = 0.20 \checkmark$$

### In the code

```python
mu_annual = 0.05
sigma_annual = 0.20
n = 100           # steps per year
delta = 1 / n

mu = mu_annual * delta          # 0.0005
sigma = sigma_annual * np.sqrt(delta)   # 0.02
```

So your original `mu=0.001, sigma=0.02` corresponds to roughly **10% annual drift** and **20% annual volatility** at 100 steps per year.
